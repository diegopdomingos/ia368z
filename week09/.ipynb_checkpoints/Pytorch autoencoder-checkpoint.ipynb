```python
import torch
import torchvision
import numpy as np
from PIL import Image
from IPython import display
from matplotlib.pyplot import imshow
import torch
from sklearn.datasets import load_iris
from torch.autograd import Variable
import torch.nn.functional as F
import torch.nn as nn
import torch.utils.data
import torch.optim as optim
from torchvision import datasets, transforms
from itertools import count
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Some constants
BATCH_SIZE = 50
TOTAL_EPOCHS = 25
TINY_DATASET = False
```

```python
data_train = torchvision.datasets.MNIST('/data/datasets/mnist_pytorch/', train=True)
data_test  = torchvision.datasets.MNIST('/data/datasets/mnist_pytorch/', train=False)
```

```python
data_train.train_data.shape, data_test.test_data.shape
```

```python
for k,(x,y) in enumerate(data_test):
    display.display(x)
    print('label=',y)
    if k > 2:
        break
```

```python
if TINY_DATASET:
    n_train_samples = 1000
    n_test_samples = 500
    data_train.train_data = data_train.train_data[:n_train_samples]
    data_test.test_data   = data_test.test_data[:n_test_samples]
    
# Load the data set as DataLoader
X_train = data_train.train_data.type(torch.FloatTensor) / 255.
y_train = data_train.train_labels.type(torch.LongTensor)


X_test = data_test.test_data.type(torch.FloatTensor) / 255.
y_test = data_test.test_labels.type(torch.LongTensor)

# Convert to dataloader
train_set = torch.utils.data.TensorDataset(X_train, y_train)
dl_train = torch.utils.data.DataLoader(train_set, batch_size=128, shuffle=True)

test_set = torch.utils.data.TensorDataset(X_test, y_test)
dl_test = torch.utils.data.DataLoader(test_set, batch_size=128, shuffle=False)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)
        self.encoded = nn.Linear(128, 2)
        self.fc2 = nn.Linear(2, 128)
        self.fc3 = nn.Linear(128,28*28)
    
    
    def encoder(self,x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = self.encoded(x)
        return x
    
    def decoder(self,x):
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
    def forward(self, x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = self.encoded(x)
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
# Main cell
model = Net()

loss_fn = torch.nn.MSELoss()

# initiate RMSprop optimizer
#optimizer = optim.SGD(model.parameters())
optimizer = optim.Adam(model.parameters())
#criterion = nn.CrossEntropyLoss()
```

```python
def train(epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(dl_train):
        data, target = Variable(data), Variable(target)
        optimizer.zero_grad()
        output = model(data)
        loss = loss_fn(output, data)
        loss.backward()
        optimizer.step()
        #print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                #epoch, batch_idx * len(data), len(dl_train.dataset),
                #100. * batch_idx / len(dl_train), loss.data[0]))

def test(epoch):
    model.eval()
    test_loss = 0
    correct = 0
    for data, target in dl_test:
        data, target = Variable(data, volatile=True), Variable(target)
        output = model(data)
        test_loss += loss_fn(output, data).data[0] # sum up batch loss

    test_loss /= len(dl_test.dataset)
    print('\nTest set(epoch {}): Average loss: {:.6f}\n'.format(epoch,test_loss))
```

```python
# Training
for epoch in range(1, TOTAL_EPOCHS + 1):
    train(epoch)
    test(epoch)
```

```python
# Convert to numpy, so we can visualize it in matplotlib
x_test = X_test.numpy()

# Set as eval model and print examples before and after compression
# in the training data set
model.eval()
output = model(Variable(dl_test.dataset.data_tensor))
output = output.data.numpy()

plt.figure(figsize=(20,4))
plt.gray()

num_examples = 10

for i in range(1,num_examples):
    
    # Before
    ax = plt.subplot(2,num_examples,i)
    plt.imshow(x_test[i])
    
    # After
    ax = plt.subplot(2,num_examples,num_examples+i)
    plt.imshow(output[i].reshape(28,28),cmap="gray")
```

```python
# Check how our test dataset is distributed in a 2 dimension visualization
model.eval()

output = model.encoder(Variable(dl_test.dataset.data_tensor))
output = output.data.numpy()

plt.figure(figsize=(7, 6))
plt.scatter(output[:, 0], output[:, 1], c=y_test.numpy(), cmap=plt.cm.Paired_r)
plt.title('Espaço de teste')
plt.colorbar()
plt.show()
```
