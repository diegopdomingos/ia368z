```python
POLY_DEGREE = 1
W_target = torch.randn(POLY_DEGREE, 1) * 5
b_target = torch.randn(1) * 5


# Load iris dataset
iris = load_iris()

data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

# Load X_b and y sets
X_b = data[:,0:1]
y = data[:,1:2]

# We must concatenate the X_b with ones column to match the
# multiplicator factor of wT[0]
X_b = np.concatenate((np.ones((X_b.shape[0],1)),X_b),axis=1)

X_b = torch.from_numpy(X_b)
y = torch.from_numpy(y)


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 1)
        #self.fc2 = nn.Linear(3,1)
        
    def forward(self,x):
        x = self.fc1(x)
        #x = self.fc2(x)
        return x
        return F.log_softmax(x)
    
model = Net()

optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
criterion = torch.nn.MSELoss(size_average=False)

data, target = Variable(X_b.float()), Variable(y.float(),requires_grad=False)

for i in range(500):

    y_pred = model(data)

    loss = criterion(y_pred,target)
    print(i,loss.data[0])

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print(y_pred)
print(target)
```
