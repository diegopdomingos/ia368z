```python
import torch
from sklearn.datasets import load_iris
from torch.autograd import Variable
import torch.nn.functional as F
import torch.nn as nn
import torch.utils.data
import torch.optim as optim
from torchvision import datasets, transforms
from itertools import count
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
X = iris.data[:,::2]  # comprimento das sépalas e pétalas, indices 0 e 2
Xc = X - X.min(axis=0)
Xc /= Xc.max(axis=0)

y = torch.from_numpy(iris.target)
Xc = torch.from_numpy(Xc)
#y = torch.from_numpy(y_o.reshape(150,1))
#y_oh = torch.zeros(y.size(0), 3).scatter_(1,y,1)

random = np.random.choice(len(Xc), len(Xc), replace=False)

X_train = Xc[torch.LongTensor(random[0:130])]
y_train = y[torch.LongTensor(random[0:130])]
X_test = Xc[torch.LongTensor(random[130:])]
y_test = y[torch.LongTensor(random[130:])]
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 3)
        
    def forward(self, x):
        x = self.fc1(x)
        return F.log_softmax(x)
```

```python
model = Net()
```

```python
optimizer = optim.SGD(model.parameters(), lr=1e-4, momentum=0.9)
```

```python
def train(X,y,epoch,b=10):
    model.train()
    random = np.random.randint(0,len(X),size=(1,b))
    data, target = Variable(X[torch.LongTensor(random[0])].float()), Variable(y[torch.LongTensor(random[0])])
    optimizer.zero_grad()
    output = model(data)
    loss = F.nll_loss(output, target)
    loss.backward()
    optimizer.step()
    print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
            epoch, len(data), len(X),100. *len(data) / len(X), loss.data[0]))

def test(X,y):
    model.eval()
    test_loss = 0
    correct = 0
    data, target = Variable(X.float(), volatile=True), Variable(y)
    output = model(data)
    test_loss += F.nll_loss(output, target, size_average=False).data[0] # sum up batch loss
    pred = output.data.max(1, keepdim=True)[1] # get the index of the max log-probability
    correct += pred.eq(target.data.view_as(pred)).cpu().sum()

    test_loss /= len(X)
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(X),
        100. * correct / len(X)))
```

```python
for epoch in range(1, 500 + 1):
    train(X_train,y_train,epoch,b=50)
    test(X_test,y_test)
```
