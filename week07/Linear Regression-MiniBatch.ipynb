```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 1)
        #self.fc2 = nn.Linear(3,1)
        
    def forward(self,x):
        x = self.fc1(x)
        #x = self.fc2(x)
        return x
        return F.log_softmax(x)
```

```python
def mini_batch_gradient_descent(X, y, model,b=10, learning_rate=1e-4, n_epochs=1000):

    w1 = torch.randn(POLY_DEGREE, 1) * 5
    
    loss_fn = torch.nn.MSELoss(size_average=False)
    
    for i in range(n_epochs):
        random = np.random.randint(0,len(X),size=(1,b))
        
        y_pred = model(X[torch.LongTensor(random[0])])

        
        loss = loss_fn(y_pred, y[torch.LongTensor(random[0])])
        print(i,loss)

        
        model.zero_grad()
        
        loss.backward()
        
        for param in model.parameters():
            param.data -= learning_rate * param.grad.data
        
    return 
```
