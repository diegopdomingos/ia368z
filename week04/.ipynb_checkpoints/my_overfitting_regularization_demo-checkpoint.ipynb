# Exercício para demonstração regularização L2 e/ou L1

Com base no notebook [overfitting_regularization](https://t-adessowiki.fee.unicamp.br/user/lotufo/notebooks/lotufo/ia368z/deliver/overfitting_regularization.ipynb),
fazer uma demonstração utilizando rede neural de uma entrada para o problema de regressão, procurando mostrar o problema do overfitting quando o modelo está muito complexo (muitas camadas, muitos neurônios), porém é regularizado adicionando-se a regularização pela norma L2 e/ou L1.

## Importação

```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
np.set_printoptions(precision=3) # ponto flutuante com 3 casas para facilitar a impressão
```

## Geração do dataset

```python
#Variância do ruído
var = 0.3

#Conjunto de treino
train_size = 10
x_train = np.linspace(0,1,train_size)
y_train = np.sin(2*np.pi*x_train) + np.random.normal(0,var,train_size) #sinal + ruido

#Conjunto de teste
test_size = 100
x_test= np.linspace(0,1,test_size)
y = np.sin(2*np.pi*x_test)
y_test = y + np.random.normal(0,var,test_size) #sinal + ruido

# Gráfico do sinal sem ruído e do conjunto de treinamento gerado
plt.figure()
plt.plot(x_test,y,linewidth = 2.0,label = r'Modelo: $sin(2 \pi x)$')
plt.scatter(x_train,y_train,color='red',label = "Modelo + ruido")
plt.plot(x_test,y_test,'x',color='black',label = "Teste")
plt.legend(loc = (0.02, 0.18))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
